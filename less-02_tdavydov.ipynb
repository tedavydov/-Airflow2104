```python
import os
import pandas as pd
import numpy as np

file_path_inp = os.path.join(os.path.expanduser('~'), 'titanic.csv')

url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df0 = pd.read_csv(url)

print(file_path_inp)
df0.to_csv(file_path_inp, encoding='utf-8')

df1 = pd.read_csv(file_path_inp)
df = np.round(df1.pivot_table(index=['Pclass'], values='Fare', aggfunc='sum').reset_index(), 2)

file_path_out = os.path.join(os.path.expanduser('~'), 'titanic_mean_fares_test.csv')
print(file_path_out)
df.to_csv(file_path_out)

df.head()
```
