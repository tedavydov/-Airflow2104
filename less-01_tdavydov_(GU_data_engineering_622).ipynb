__5.__ Опишите своими словами, как Вы поняли, чем отличается task от operator?

Ответ: task - это выполняемая задача в рамках operator 
 - один оператор может содержать несколько тасков (но так делать нежелательно)

Оператор - это по сути шаблон, на основании него создаются и настраиваются звенья цепи выполнения ДАГа.

Операторов может быть много, можно создавать свои операторы - предназначенные для выполнения собственных задач.
